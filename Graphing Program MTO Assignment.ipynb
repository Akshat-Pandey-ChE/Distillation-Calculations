```python
import matplotlib.pyplot as plt
import numpy as nd
```

```python
x = nd.array(range(100))
```

```python
x = x/100
```

```python
y = 5*x/(2+ (3*x))  ##Equation of EQBM Curve obtained from relative volatility
```

```python
y1 = nd.array(range(100))
y1 = y1/100
x1 = y1  ## Equation of Ideal VLE Curve
x3 = nd.array(range(100))
x3 = x3/100
y3 = (5 - (6*x3))/4  ##Equation of Q line obtained from liquid and vapor precentage in feed given and Zf value
x4 = nd.array(range(100))
x4 = x4/100
y4 = 0.6064*x4 +0.3857    ##Equation of EOL from xd,xd and intersection point of EQBM curve and Q line
x5 = nd.array(range(100))
x5 = x5/100
y5 = 1.5901*x5 -0.0177    ## Equation of SOL from xb,xb and intersection point of EQBM curve and Q line
```

```python
plt.plot(x1,y1, label ="X = Y", color = "red")
plt.plot(x,y, label = "EQBM Curve", color = "green")
plt.plot(x3,y3, label = " Q - Line", color = "black")
plt.plot(x4,y4, label = " EOL ", color = "blue")
plt.plot(x5,y5, label = " SOL ", color = "yellow")
plt.grid()
plt.legend()
```
